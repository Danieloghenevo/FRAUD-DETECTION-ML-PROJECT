### feature engineering

```python
%%time

# calculating last digit difference
def calculate_phone_number_diff(group):
    group['mobile_number_for_mobile_money'] = pd.to_numeric(group['mobile_number_for_mobile_money'], errors='coerce')
    group['phone_number_diff'] = group['mobile_number_for_mobile_money'].diff()
    group['phone_number_diff'] = group['phone_number_diff'].fillna(0)  # Or -999 for first transaction
    return group

df1 = df1.groupby('customer_email').apply(calculate_phone_number_diff)
```

```python
df1.head()
```

```python
# AGGREGATING CUSTOMER METRICS
# - Number of unique phone numbers: how many different phone numbers each customer uses. obviously, a high count is suspicious.
# - Number of txns with different numbers: how many txns involve a change in phone number from the previous txn
# - Proportion of txns with different numbers: dividing the number of txns with different numbers by the total number of txns for the customer
# - Average/median difference: calculating the average or median absolute difference between phone numbers (excluding 0 differences, 
#                              which are legit same-number entries). Small average differences are highly suspicious.

# calculating metrics and merging directly into df1
def calculate_customer_metrics(group):
    unique_phone_numbers = group['mobile_number_for_mobile_money'].nunique()
    diff_number_txns = (group['phone_number_diff']!= 0).sum()
    total_transactions = len(group)
    proportion_different = diff_number_txns / total_transactions if total_transactions > 0 else 0
    differences = group['phone_number_diff'].abs()
    differences = differences[differences!= 0]
    average_difference = differences.mean() if not differences.empty else 0

    return pd.Series({
        'unique_phone_numbers': unique_phone_numbers,
        'diff_number_txns': diff_number_txns,
        'total_transactions': total_transactions,
        'proportion_different_numbers': proportion_different,
        'average_phone_number_difference': average_difference
    })
```

```python
df2 = df1.copy()
df2 = df2.rename(columns={'customer_email': 'cust_email'}) # renaming customer_email to avoid errors
df2 = df2.reset_index()
df2 = df2.join(df2.groupby('customer_email').apply(calculate_customer_metrics), on='customer_email', how='left')
```

```python
df2.head()
```

```python
# Identify Shared Phone Numbers with Different Customer Info (Binary Flag)
shared_phone_numbers = df2.groupby('mobile_number_for_mobile_money', as_index=False).filter(lambda x: len(x) > 1)
shared_phone_numbers = shared_phone_numbers.groupby('mobile_number_for_mobile_money', as_index=False).agg({'customer_email': 'nunique'}) # Count unique customer_ids for each phone
shared_phone_numbers = shared_phone_numbers[shared_phone_numbers['customer_email'] > 1] #Keep phone numbers shared by more than one customer
shared_phone_numbers['shared_phone_number_flag'] = 1  # Create the binary flag
shared_phone_numbers = shared_phone_numbers[['mobile_number_for_mobile_money', 'shared_phone_number_flag']] #Keep only the phone number and the shared_phone_number_flag

df2 = pd.merge(df2, shared_phone_numbers, on='mobile_number_for_mobile_money', how='left')
df2['shared_phone_number_flag'] = df2['shared_phone_number_flag'].fillna(0)
```

```python
# frequency of txns per customer
df2['txn_count_per_customer'] = df2.groupby('customer_email')['transaction_reference'].transform('count')
```

```python
# frequency of txns per customer per hour
df2['txn_count_per_customer_per_hour'] = df2.groupby(['customer_email', 'created_hod'])['transaction_reference'].transform('count')
```

```python
# maximum transactions within an hour (per day)
# df2.drop(columns=['max_txns_in_an_hour'], inplace=True)
max_txns_per_hour_per_customer = df2.groupby(['customer_email', 'created_dow', 'created_hod'])['transaction_reference'].count().groupby(['customer_email', 'created_dow']).max()
df2 = df2.merge(max_txns_per_hour_per_customer.rename('max_txns_in_an_hour_per_customer'), left_on=['customer_email', 'created_dow'], right_index=True, how='left')
```

```python
# frequency of txns per customer per day
df2['txn_count_per_customer_per_day'] = df2.groupby(['customer_email', 'created_dow'])['transaction_reference'].transform('count')
df2.head()
```

```python
# checking out phone number pattern + high transaction amount
def check_phone_amount(group):
    group['phone_txn_amount_suspicious'] = 0
    for customer_email in group['customer_email'].unique():
        customer_group = group[group['customer_email'] == customer_email]
        phone_numbers = customer_group['mobile_number_for_mobile_money'].unique()
        if len(phone_numbers) > 1: #Check if the customer has more than one phone number
            high_amounts = customer_group[customer_group['amount'] > customer_group['amount'].quantile(0.90)] #Check for high transaction amounts
            if not high_amounts.empty: #If there are high transaction amounts, flag them
                group.loc[high_amounts.index, 'phone_txn_amount_suspicious'] = 1
    return group
df2 = df2.groupby('customer_email').apply(check_phone_amount).reset_index(drop=True)
```

```python
df2.phone_txn_amount_suspicious.value_counts()
```

```python
# sudden spikes in txn count per customer
# Here, I'm using a rolling window (7 days) to calculate a moving average of txn counts per customer.
# The window size can be adjusted as needed. The spike flag is set if the current txn count is significantly 
# higher (I set it to 2.5x here) than the rolling average.
df2['txn_count_per_customer'] = df2.groupby('customer_email')['transaction_reference'].transform('count')
df2['rolling_txn_count_per_customer'] = df2.groupby('customer_email')['transaction_reference'].transform(lambda x: x.rolling(window=7, min_periods=1).count()) # 7-day rolling window
df2['txn_spike_flag'] = (df2['txn_count_per_customer'] > 2.5 * df2['rolling_txn_count_per_customer']).astype(int) # Example: 2.5x the rolling average
```

```python
df2.txn_spike_flag.value_counts()
```

```python
# unusual txn amounts per customer
# flagging txns that are 5x higher than a customer's average txn for cases where a customer's total txn count is >=10
df2['avg_txn_amount_per_customer'] = df2.groupby('customer_email')['amount'].transform('mean')
# Identify transactions where total_transactions > 9
total_txn_threshold = df2['total_transactions'] > 9
df2['unusual_txn_flag'] = 0
df2.loc[total_txn_threshold, 'unusual_txn_flag'] = (df2.loc[total_txn_threshold, 'amount'] > 
                                                    5 * df2.loc[total_txn_threshold, 'avg_txn_amount_per_customer']).astype(int)
```

## Considerations for feature engineering

- drop merchant_bears_cost
- encode payment_description using a target encoding approach, where the encoding reflects the likelihood of fraud for each description.
- create a binary flag (is_suspicious_description) for transactions with these two suspicious payment descriptions.
- extract the day of the week as a categorical feature (txn_day_of_week) and one-hot encode it.
- add a feature (is_monday) to flag transactions that occur on Monday.
- Extract the hour of the day as a numerical feature (txn_hour).
- Add a binary flag (is_peak_hour) for transactions occurring at 10 AM.
- Extract the month as a categorical feature (txn_month) and one-hot encode it.
- Add a binary flag (is_december) for transactions in December.
- Use one-hot encoding for payment_source_type.
- Create a binary flag (is_bank_transfer) for bank transfer transactions.
- Encode the merchant as a categorical variable (merchant) with one-hot encoding or target encoding.
- Add a feature (merchant_risk_level) based on risk levels: high, above average, and low.
- Create binary flags for is_merchantA and is_merchantB.
- Encode the processor (processor) as a categorical feature with one-hot encoding.
- Add a binary flag (is_processorA) to identify transactions processed by ProcessorA.
- One-hot encode payment_status.
- Add a binary flag (is_successful) for transactions marked as successful.
- Create a feature (customer_name_count) to record the number of occurrences of a customer name for the merchant.
- Add a binary flag (is_suspicious_name) for names appearing frequently in fraud transactions.
- Despite the lack of statistical significance, retain payment_completion_time_secs as a numerical feature. It might contribute in combination with other features.
